```python
def count_by(x, n):
    # x  o valor e n  o nmero de mltiplos desse valor, deve ser retornado em uma array
    c = []
    y = 0
    for i in range(0, n):
        y = y + 1 
        c.append(x * y)
    return c
```

### My code:

```python
def positive_sum(arr):
    y = 0
    for i in range(len(arr)):
        if arr[y] < 0:
            arr.pop(y)
        else:
            y = y + 1
    return sum(arr)
```

### My code:

```python
def find_needle(haystack):
    y = 0
    for i in range(len(haystack)):
        if haystack[y] == 'needle':
            return f"found the needle at position {y}"
        else:
            y = y + 1
```

### My code:

```python
def past(h, m, s):
    # 1 segundo tem 1000 milissegundos
    # 1 minuto tem 60 segundos == 60000
    # 1 hora tem 60 minutos == 60000 * 60
    h = h * (60000 * 60)
    m = m * 60000
    s = s * 1000
    return h + m + s
```

```python
def reverse_seq(n):
    x = []
    x.append(n)
    for i in range(1, n):
        n = n - 1
        x.append(n)
    return(x)
```

```python
def abbrev_name(name):    
    fl = name[0]
    sl = name.find(' ') + 1
    return (fl + '.' + name[sl]).upper()
```
